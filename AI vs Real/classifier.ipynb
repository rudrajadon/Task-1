```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define ImageDataGenerator with validation split
train = ImageDataGenerator(validation_split=0.2, rescale=1./255)

# Training dataset (80% of the data)
train_dataset = train.flow_from_directory(
    '/Users/rudrajadon/Desktop/X/TasksML/Data/Train/',
    target_size=(200, 200),
    batch_size=32,
    class_mode='binary',
    subset='training', 
    seed=42            
)

# Validation dataset (20% of the data)
validation_dataset = train.flow_from_directory(
    '/Users/rudrajadon/Desktop/X/TasksML/Data/Train/',
    target_size=(200, 200),
    batch_size=32,
    class_mode='binary',
    subset='validation', 
    seed=42            
)
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16,(3,3),activation = 'relu',input_shape =(200,200,3)),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(32,(3,3),activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(64,(3,3),activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    ##
                                    tf.keras.layers.Flatten(),
                                    ##
                                    tf.keras.layers.Dense(512,activation= 'relu'),
                                    ##
                                    tf.keras.layers.Dense(1,activation='sigmoid')
                                   ])
```

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(learning_rate=0.001),
              metrics=['accuracy'])
```

```python
model_fit = model.fit(train_dataset,
                      steps_per_epoch = 10,
                      epochs = 30,
                      validation_data = validation_dataset)
```

```python
from keras.preprocessing import image
import os
import numpy as np
import pandas as pd
import random

dir_path = "/Users/rudrajadon/Desktop/X/TasksML/Data/Test"
image_ids = []
real_or_ai = []

# Define acceptable image extensions
valid_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.gif']

for i in os.listdir(dir_path):
    # Check if the file is a valid image based on its extension
    if any(i.lower().endswith(ext) for ext in valid_extensions):
        try:
            img = image.load_img(os.path.join(dir_path, i), target_size=(200, 200))
            X = image.img_to_array(img)
            X = np.expand_dims(X, axis=0)

            # Predicting the image
            val = model.predict(X)  # Assuming model output is a probability
            if val[0] < 0.5:  # 0 represents 'AI' class
                image_ids.append(i)
                real_or_ai.append("AI")
            else:
                image_ids.append(i)
                real_or_ai.append("Real")
        except Exception as e:
            print(f"Error loading image {i}: {e}")
            image_ids.append(i)
            real_or_ai.append(random.choice(["Real", "AI"]))
    else:
        print(f"Skipping non-image file {i}")
        image_ids.append(i)
        real_or_ai.append(random.choice(["Real", "AI"]))

# Creating a DataFrame
image_df = pd.DataFrame({"Id": image_ids, "Label": real_or_ai})
image_df['Id'] = image_df['Id'].str.extract('(\d+)').astype(int)

# Sorting the DataFrame by the 'Id' column
image_df_sorted = image_df.sort_values(by='Id')
image_df_sorted['Id'] = 'image_' + image_df_sorted['Id'].astype(str)

# Printing the DataFrame
print(image_df_sorted[['Id', 'Label']].to_string(index=False))

# Saving the DataFrame to CSV file
output_csv_path = "/Users/rudrajadon/Desktop/X/TasksML/Data/predictions5.csv"
image_df_sorted[['Id', 'Label']].to_csv(output_csv_path, index=False)

print("CSV file saved")
```
